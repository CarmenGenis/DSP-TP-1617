## Exercise 1 - Causality

 **1.1** Considering the system defined by the equation $y_k=(x_k+x_{k+1})/2$, check its causality property by examining the response to the signal $H(k-4)$ or `step(4,N)`.

```python
#Function Step

def Step (n=0, N=20):
    
    if n > (N-1):
        print ('n is out of the range') 
        
    H = np.zeros (N, ) # Creating an array 0
    
    for a in np.arange (n,N):
        H[a]=1
        

        
    return H
```

```python



N=10
x = Step(4,N) 
       
y = np.zeros (N, ) # Creating an array 0
    
for k in np.arange (0,N-1):
    y[k] = (x[k] + x[k+1])/2

plt.stem(y)
```

**1.2** Propose a modification to obtain a causal version

```python
N=10
x = Step(4,N)
       
y = np.zeros (N, ) # Creating an array 0
#Like the previous function is based in the future step (k+1) our Graphic is going to star in 3 which mean that its no causal    
for k in np.arange (1,N): 
    y[k] = (x[k] + x[k-1])/2 #Here we take the previous step and the range of above we have to modify as well
                             #Now the system is causal.
plt.stem(y)
```
